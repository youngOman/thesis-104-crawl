### 薪資資料預處理

```python
def encode_salary(salary_low, salary_high):
    # 計算薪資範圍的平均值
    average_salary = (salary_low + salary_high) / 2
    # 根據行政院總計處提供的十分位數進行編碼
    if 27300 <= average_salary < 32600:
        return 1
    elif 32600 <= average_salary < 39000:
        return 2
    elif 39000 <= average_salary < 45500:
        return 3
    elif 45500 <= average_salary < 54500:
        return 4
    elif 54500 <= average_salary < 62100:
        return 5
    elif 62100 <= average_salary < 74200:
        return 6
    elif 74200 <= average_salary < 93000:
        return 7
    elif 93000 <= average_salary < 122000:
        return 8
    elif average_salary >= 122000:
        return 9
    else:
        return 3  # 對於不符合以上條件的情況，可以返回 None 或其他預設值

# 假設您的數據集中最低薪資和最高薪資的列名分別為 '最低薪資' 和 '最高薪資'
data['薪資範圍編碼'] = data.apply(lambda row: encode_salary(row['最低薪資'], row['最高薪資']), axis=1)

# # 首先，計算並儲存平均薪資
# data['average_salary'] = (data['最低薪資'] + data['最高薪資']) / 2
# # 顯示包含計算後的「平均薪資」前 15筆
# data[['最低薪資', '最高薪資', 'average_salary']].head(15)

# 查看結果
data[['最低薪資', '最高薪資', '薪資範圍編碼']].head(15)


# 假設 NaN 值用特定整數替代，例如 -1
data['薪資範圍編碼'] = data['薪資範圍編碼'].fillna(-1)
# 現在將浮點數轉換為整數
data['薪資範圍編碼'] = data['薪資範圍編碼'].astype(int)

# 查看結果
data[['最低薪資', '最高薪資', '薪資範圍編碼']].head()
```

```python
# 計算薪資範圍的平均值並轉換為整數
def compute_average_salary(salary_low, salary_high):
    # 計算薪資範圍的平均值
    average_salary = (salary_low + salary_high) / 2
    # 返回去除小數點的薪資
    return int(average_salary)

# 計算每一筆資料的平均薪資並轉換為整數，直接存儲到新的列 '薪資'
data['薪資'] = data.apply(lambda row: compute_average_salary(row['最低薪資'], row['最高薪資']), axis=1)

# 查看結果
data[['薪資']].head()
```

### 學歷資料預處理

```python
def encode_education_level(education):
    education_mapping = {'不拘': 1, '高中': 1, '專科': 2, '大學': 3, '碩士': 4, '博士': 5}
    for level in education_mapping:
        if level in str(education):
            return education_mapping[level]
    return None  # 如果沒有匹配的學歷等級，則返回 None 或者您可以選擇其他的預設值

# 應用函數並創建一個新列
data['學歷要求編碼'] = data['學歷要求'].apply(encode_education_level)

# 查看結果
data[['學歷要求', '學歷要求編碼']].head()
```

### 工作年資資料預處理

```python
def encode_work_experience(experience):
    if experience == '不拘':
        return 1
    else:
        # 使用split()函數分割字符串，並嘗試提取第一個數字
        # 如果分割後的第一部分能夠轉換為數字，則直接使用該數字作為編碼
        # 如果包含"以上"，則編碼為 5
        parts = experience.split('年')
        try:
            years = int(parts[0])  # 提取年份
            if '以上' in experience and years > 4:
                return 5  # 對於'X年以上'的情況，編碼為 5
            return min(years, 5)  # 確保返回值不超過 5
        except ValueError:
            # 轉換失敗，分割的第一部分不是一個數字
            return 1

data['工作經驗編碼'] = data['工作經歷要求'].apply(encode_work_experience)

data[['工作經歷要求', '工作經驗編碼']].head(5)
```

### 是否出差資料預處理

```python
def encode_travel_requirement(requirement):
    if pd.isna(requirement):
        return 0  # 或者根據業務規則選擇合適的填充值
    elif requirement.startswith('無'):
        return 0
    elif requirement.startswith('需'):
        return 1
    else:
        return None  # 非預期的值

# 應用函數並創建一個新列
data['是否需出差外派編碼'] = data['是否需出差外派'].apply(encode_travel_requirement)

# 查看結果
data[['是否需出差外派', '是否需出差外派編碼']].head(5)
```

### 員工人數 (公司規模)

```python
import pandas as pd
import re

def classify_company_size(employees):
    match = re.search(r'\d+', str(employees))
    if match:
        number_of_employees = int(match.group(0))
        if number_of_employees < 100:
            return 1  # 小型企業
        elif 100 <= number_of_employees < 200:
            return 2  # 中型企業
        else:
            return 3  # 大型企業
    else:
        return 1 

data['公司規模'] = data['員工人數'].apply(classify_company_size)

data[['員工人數', '公司規模']].head()
```

### 職缺描述、其他條件 文字主觀性、可讀性

```python
import pandas as pd
from textblob import TextBlob
from textatistic import Textatistic

# Define a function to analyze text and calculate readability and subjectivity
def analyze_text(data):
    # Initialize new columns to ensure we do not overwrite or delete original columns
    data['職務內容可讀性'] = 0.0
    data['職務內容主觀性'] = 0.0
    data['其他條件可讀性'] = 0.0
    data['其他條件主觀性'] = 0.0

    # 確保文本列是字符串類型，並處理 NaN值 *** 這段可以加到論文，說我們是如何處理空值***
    data['職務內容'] = data['職務內容'].fillna('').astype(str)
    data['其他條件'] = data['其他條件'].fillna('').astype(str)
    
    for i in range(len(data)):
        # Process job content
        job_content = data.loc[i, '職務內容']
        if job_content:
            job_blob = TextBlob(job_content)
            data.at[i, '職務內容主觀性'] = job_blob.sentiment.subjectivity
            try:
                readability_stats = Textatistic(job_content).scores
                data.at[i, '職務內容可讀性'] = readability_stats['flesch_score']
            except:
                data.at[i, '職務內容可讀性'] = None  # Some texts may be too short to calculate readability

        # Process other conditions
        other_conditions = data.loc[i, '其他條件']
        if other_conditions:
            other_blob = TextBlob(other_conditions)
            data.at[i, '其他條件主觀性'] = other_blob.sentiment.subjectivity
            try:
                readability_stats = Textatistic(other_conditions).scores
                data.at[i, '其他條件可讀性'] = readability_stats['flesch_score']
            except:
                data.at[i, '其他條件可讀性'] = None

    return data

# Call the function to perform text analysis
analyzed_data = analyze_text(data)
```

### 缺失值插補

```python
# 計算可讀性的平均值
mean_readability = data['其他條件可讀性'].mean()

# 用平均值填充「其他條件可讀性」的缺失值
data['其他條件可讀性'].fillna(mean_readability, inplace=True)
```

```python
# 計算可讀性的平均值
mean_readability = data['職務內容可讀性'].mean()

# 用平均值填充「其他條件可讀性」的缺失值
data['職務內容可讀性'].fillna(mean_readability, inplace=True)
```
